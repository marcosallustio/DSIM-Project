```python
import os
import time
import uuid
import cv2
```

```python
IMAGES_PATH = "/Users/sallu/Desktop/foto/"
inizio = 100
number_images = 400
```

```python
cap = cv2.VideoCapture(0)

for imgnum in range(inizio, inizio + number_images):
    print('Collecting image {}'.format(imgnum))
    ret, frame = cap.read()

    imgname = os.path.join(IMAGES_PATH, "Salvatore_{}.jpg".format(imgnum))
    cv2.imwrite(imgname, frame)
    cv2.imshow('frame', frame)
    cv2.waitKey(500)  # Attendiamo 500 millisecondi (0.5 secondi)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
